#7)Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)

```python
import scipy.stats as stats
# One-way ANOVA
f_stat, p_value = stats.f_oneway(df['etest_p'], df['mba_p'])
print("F-statistic:", f_stat)
print("p-value:", p_value)
```

#Ans: p_value should be less than 0.05. here the p-value (4.67) is more than 0.05.So it is not rejecting the null hypothesis.since there is a no 
significant difference between etest_p and mba_p.

#8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('salary ~ C(degree_t) + C(specialisation) + C(degree_t):C(specialisation)', data=df).fit()
sm.stats.anova_lm(model, typ=2)
```

#Ans: I used two way anova, to check the similarity between 2 factors (degree_t,specialisation) as indep and salary as dep. According to the report, degree_t p_value (0.2) >0.05,it is not rejecting the null hypothesis, there is no significance difference. specialisation p_value(0.00005) < 0.05, its rejecting the null hypothesis and there is a significant difference. similarity between the degree_t and specialisation p_value (0.78)>0.05,it is not rejecting the null hypothesis and there is a no significant difference between them.

#9)Convert the normal distribution to standard normal distribution for salary column

```python
mean_salary = df['salary'].mean()
std_salary = df['salary'].std()

# Standardize (z-score)
df['salary_z'] = (df['salary'] - mean_salary) / std_salary

# Check the result
print(df[['salary', 'salary_z']].head())
```

#11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

```python
from scipy.stats import ttest_rel
sci_tech = df[ df['degree_t'] == 'Sci&Tech']['etest_p']
sci_tech1 = df[ df['degree_t'] == 'Sci&Tech']['mba_p']
ttest_rel(sci_tech,sci_tech1)
```

#Ans: At the 5% significance level, the p-value is less than 0.05. Hence, we reject the null hypothesis. There is a significant difference between etest_p and mba_p for Sci&Tech students.
