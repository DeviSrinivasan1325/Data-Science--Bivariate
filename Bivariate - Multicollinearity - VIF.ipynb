```python
#Variance inflation factor
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return (vif)
    
```

```python
calc_vif(dataset[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']])
```

```python
# VIF explaintion
ssc_p - 66.94 - Severe multicollinearity  
hsc_p - 56.08 - Severe multicollinearity  
degree_p - 112.70 - Extreme multicollinearity 
etest_p - 32.67  Severe multicollinearity  
mba_p - 99.28 - Severe multicollinearity 
```
